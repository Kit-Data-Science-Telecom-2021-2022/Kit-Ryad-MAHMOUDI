```python
# import des modules usuels
import numpy as np
import pandas as pd
import json
from shapely.geometry import Polygon
import matplotlib.pyplot as plt
# options d'affichage
pd.set_option("display.min_rows", 16)
```

```python
# chargement et traitement des données
geo = pd.read_csv("correspondance-code-insee-code-postal.csv",
                   sep=';',
                   usecols=range(11),
                   index_col="Code INSEE")
geo = geo.sort_index()
geo[['Latitude', 'Longitude']] = geo["geo_point_2d"].str.extract("(.*), (.*)").astype(float)
geo.head()
```

**Projet maison n° 1**

- Ajoutez une colonne 'CP Ville' avec le Code postal + un espace + et le nom de la Ville.
- Ecrivez une fonction qui détermine la commune la plus proche d'un point à partir de sa latitude et sa longitude.
- Ajoutez une fonction de conversion pour pouvoir utiliser la première fonction avec un GPS (degrés, minutes, secondes).

### 

### Ajout de la colonne 'CP Ville'

```python
geo['CP Ville'] = geo['Code Postal'] + ' ' + geo['Commune']
geo['CP Ville'].iloc[4]
geo.shape[0]
```

### Fonction recherche de ville

```python
def ville(lat, long):
    dist={}
    for i in range(geo.shape[0]):
        lati = geo['Latitude'].iloc[i]
        longi = geo['Longitude'].iloc[i]
        d=np.sqrt((lati-lat)**2+(longi-long)**2)
        dist[geo['CP Ville'].iloc[i]] = d
    
    return min(dist, key=dist.get)
```

```python
# conversion degrés, minutes, secondes => décimal
def dms2dec(deg, mn, sec):
    return deg + mn / 60 + sec / 3600
```

```python
# on applique la fonction à une coordonnée tirée au hasard
# init du random
np.random.seed(0)

a, b = 41.5, 51.1  # latitude min et max de la France métropolitaine
lat = (b - a) * np.random.random() + a
a, b = -5.1, 9.5  # longitude min et max de la France métropolitaine
long = (b - a) * np.random.random() + a

ville(lat, long)
```

```python
# à partir de coordonnées GPS précises
ville(dms2dec(48, 42, 52), dms2dec(2, 14, 45))
```

### Chargement des json en dictionnaires

```python
geo["geo_shape"]=geo["geo_shape"].apply(lambda x:json.loads(x))
```

### Décompte des valeurs accédés avec la clé "type"

```python
geo["geo_shape"].apply(lambda x: x["type"]).value_counts()
```

### Décompte des longueurs des listes accédées avec la clé "coordinates"

```python

print('Décompte des longueurs des listes accédées avec la clé "coordinates" : \n')
print(geo["geo_shape"].apply(lambda x: len(x["coordinates"])).value_counts())

Nb_polygone_max =geo["geo_shape"].apply(lambda x: len(x["coordinates"])).value_counts().index[-1]
Nb_polygone_max2 =geo["geo_shape"].apply(lambda x: len(x["coordinates"])).value_counts().index[-2]

print("\n\nLa ville la plus complexe possède ",Nb_polygone_max, " polygones")
print("La deuxième ville la plus complexe possède ",Nb_polygone_max2, " polygones")
```

### Commune la plus complexe géométriquement

La commune la plus complexe géométriquement est celle ayant le plus grand nombre de polygones, on sait qu'il n'y a qu'une seule ville avec 4 polygones. Cherchons la :

```python
print("La ville la plus complexe est : ")
print(geo.loc[geo["geo_shape"].apply(lambda x: len(x["coordinates"])) == Nb_polygone_max,"CP Ville"].values[0])
```

### Seconde Commune la plus complexe géométriquement

La seconde commune la plus complexe géométriquement est celle ayant le deuxième plus grand nombre de polygones, on sait qu'il n'y a qu'une seule ville avec 3 polygones après Hyères qui en contient 4. Cherchons la :

```python
print("La deuxième ville la plus complexe est : ")
print(geo.loc[geo["geo_shape"].apply(lambda x: len(x["coordinates"])) == Nb_polygone_max2,"CP Ville"].values[0])
```

### Villes de type "Polygon" mais dont la longueur des listes accédées avec la clé "coordinates" vaut 2

```python
geo.loc[(geo["geo_shape"].apply(lambda x: x["type"]) == "Polygon") & (geo["geo_shape"].apply(lambda x: len(x["coordinates"])) == 2),"CP Ville"]
```

```python
#Sauvegarde des coordonnees de ces villes dans une Series
coord_villes = geo.loc[(geo["geo_shape"].apply(lambda x: x["type"]) == "Polygon") & (geo["geo_shape"].apply(lambda x: len(x["coordinates"])) == 2),"geo_shape"].apply(lambda x: x["coordinates"])
```

### Shapely

On crée une fonction qui à partir des coordonnes vérifie qu'un polygone est inclus dans l'autre, affiche un message et trace les polygones avec Matplotlib (on trace si et seulement si le paramètre plot est égal a True).
On fait appel à la fonction pour chacune des 10 villes.


```python
def p_contient(coordonnees,nom_ville , plot):
    Polygon1 = Polygon(coordonnees[0])
    Polygon2 = Polygon(coordonnees[1])
    if Polygon1.contains(Polygon2)==True:
        if plot == False:
            BOLD = '\033[1m'
            print(BOLD + "Il y a bien un polygone inclus dans l'autre dans la ville de ",nom_ville)

        if plot == True:
            fig = plt.figure(figsize=(8,6))
            #tracé des polygones avec Matplotlib
            x1,y1 = Polygon1.exterior.xy
            x2,y2 = Polygon2.exterior.xy
            plt.plot(x1,y1,label = "Polygone 1")
            plt.plot(x2,y2,label = "Polygone 2")
            plt.legend()
            plt.title("Polygones de la ville de "+nom_ville)
            plt.show()
            print("\n")
    return 
```

```python
#Affichage et vérification des enclaves des polygones pour les 10 villes
for i in range(coord_villes.size):
    p_contient(coord_villes[i],geo.loc[coord_villes.index[i],"Commune"],False)
```

```python
#Tracé des polygones pour chaque ville
for i in range(coord_villes.size):
    p_contient(coord_villes[i],geo.loc[coord_villes.index[i],"Commune"],True)
```
